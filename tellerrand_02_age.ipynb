# does "age" has an effect on "survived"?

```python
age_data = data_train[["Age", "Survived"]].dropna(axis=0)
age_data.shape
```

```python
print(age_data.Age.min())
print(age_data.Age.max())
```

```python
corr = age_data.corr(method="pearson").abs()
sns.heatmap(corr, vmin=0, vmax=1, cmap="binary", annot=True, linewidths=0.66);
```

```python
corr = age_data.corr(method="spearman").abs()
sns.heatmap(corr, vmin=0, vmax=1, cmap="binary", annot=True, linewidths=0.66);
```

```python
pd.pivot_table(age_data, index="Survived", values = "Age", aggfunc=["mean", "median"]).round(2)
```

```python
sns.histplot(x="Age", data=age_data, stat="percent", bins=16, binrange=[0, 80], color="k");
```

```python
sns.histplot(x="Age", hue="Survived", data=age_data, stat="percent", bins=16, binrange=[0, 80], color="k");
```
